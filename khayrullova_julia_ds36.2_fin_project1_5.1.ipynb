# Сборный проект 1: исследование продаж компьютерных игр

Перед нами исторические данные из открытых источников о продажах игр до 2016 года, оценки пользователей и экспертов, жанры и платформы. Мы планируем кампанию на 2017-й год для интернет-магазина «Стримчик», который продаёт по всему миру компьютерные игры. Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании. 
<br>Данные за 2016 год могут быть неполными.

<font size=1>В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».</font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import datetime as dt
import seaborn as sns
from scipy import stats as st
```

```python
try:
    df = pd.read_csv('games.csv')
except:
    df = pd.read_csv('/datasets/games.csv')
```

Таблица имеет 11 столбцов и 16715 строк.

Согласно документации:<br>
`Name` — название игры;<br>
`Platform` — платформа;<br>
`Year_of_Release` — год выпуска;<br>
`Genre` — жанр игры;<br>
`NA_sales` — продажи в Северной Америке (миллионы проданных копий);<br>

`JP_sales` — продажи в Японии (миллионы проданных копий);<br>

`Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

<br>Названия столбцов нужно привести к `snake_case`.

<br>Оценку пользователей нужно привести к числовому типу.

## Предобработка данных

### Названия столбцов

```python
df = df.rename(columns={
    'Name':'name',
    'Platform':'platform',
    'Year_of_Release':'year_of_release',
    'Genre':'genre',
    'Other_sales':'other_sales',
    'Critic_Score':'critic_score',
    'User_Score':'user_score_10',
    'Rating':'rating'
})
df.columns   
```

Столбцы `NA_sales`, `EU_sales`, `JP_sales` не будем переименовывать. Коды стран и регионов заглавными буквами выглядят нагляднее.

### Типы данных и пропуски

```python
df['user_score_10'].value_counts(dropna=False)
```

tbd - to be determined. Значит оценок ещё нет (очень мало).
<br>Думаю, можно заменить на NaN.

```python
df.loc[df['user_score_10']=='tbd', 'user_score_10'] = None
df['user_score_10'].value_counts(dropna=False)
```

```python
df['user_score_10'] = pd.to_numeric(df['user_score_10'], errors='coerce')
df.info()
```

Заполним год выпуска средним по платформе:

```python
mean_years_dict = np.round(df.groupby('platform')['year_of_release'].mean()).to_dict()
mean_years_dict
```

```python
mean_years = df['platform'].apply(lambda x: mean_years_dict[x])
df['year_of_release'] = df['year_of_release'].fillna(mean_years)
df.info()
```

```python
df['user_score'] = df['user_score_10']*10
#df = df.convert_dtypes()
df['year_of_release'] = df['year_of_release'].astype('int')
df.info()
```

```python
df[df['name'].isna()]
```

```python
df.loc[[658,659,660,14243,14244,14245]]
```

В строках 659, 14244 нет полезной информации: ни названия, ни даже жанра, и они никак не связаны с соседними строками. Их нужно удалить.

```python
df = df.dropna(subset=['name'])
```

В остальных столбцах, где есть пропуски: `critic_score` (>50%), `user_score` (>50%), `rating` (>30%) - их очень много.
<br>Вероятно, в открытых источниках, откуда собраны данные, их не было, или не нашли, как правильно их подтянуть.
<br>Наверняка их можно было бы подтянуть по названию из каких-то ещё открытых источников, но это была бы отдельная подзадача, пока не очень представляю, как сделать (и дэдлайн). А заполнять средними/медианными/наиболее частотными значениями не представляется здесь адекватным.

### Суммарные продажи
Рассчитаем отдельный столбец с суммарными продажами по всем регионам.

```python
df['total_sales'] = df['NA_sales'] + df['JP_sales'] + df['EU_sales'] + df['other_sales']
df.head(3)
```

#### Актуальный период
Определим актуальный период на основании исследования предыдущих вопросов. Основной фактор — эти данные помогут построить прогноз на 2017 год.

```python
df_actual = df.query('year_of_release >= 2013')
```

### Продажи VS отзывы
Посмотрим, как влияют на продажи отзывы пользователей и критиков внтури одной из популярных платформ. Пусть это будет **PS4**.

Построим диаграмму рассеяния:

```python
df_ps4 = df_actual.query('platform == "PS4"')
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(2,2,1)
df_ps4.plot(kind='scatter', x='critic_score', y='total_sales', alpha=0.3, grid=True, ax=ax, logy=False);
ax = fig.add_subplot(2,2,2)
df_ps4.plot(kind='scatter', x='user_score', y='total_sales', alpha=0.3, grid=True, ax=ax, logy=False);
ax = fig.add_subplot(2,2,3)
df_ps4.plot(kind='scatter', x='critic_score', y='total_sales', alpha=0.5, grid=True, ax=ax, logy=True);
ax = fig.add_subplot(2,2,4)
df_ps4.plot(kind='scatter', x='user_score', y='total_sales', alpha=0.5, grid=True, ax=ax, logy=True);
```

Зависимость от оценок критиков явно есть. Даже скорее логарифмическая, чем линейная. Насчёт оценок пользователей пока не ясно.

Проверим корреляцию между отзывами и продажами:

```python
print('Коэф-т корреляции Пирсона для продаж и оценки пользователей:', 
      np.round(df_ps4['total_sales'].corr(df_ps4['user_score']), decimals=4))
print('Коэф-т корреляции Пирсона для продаж и оценки критиков:', 
      np.round(df_ps4['total_sales'].corr(df_ps4['critic_score']), decimals=4))
```

Действительно, зависимость от оценок критиков есть, а от оценок пользователей - не выявлена.
<br>Вероятно, для других платформ так же.

Посчитаем среднюю прибыльность по жанрам, годам выпуска.

```python
df_actual.groupby('genre')['total_sales'].mean().sort_values(ascending=False)
```

Самый прибыльный жанр - Shooter, самый неприбыльный - Adventure (тем не менее, таких игр выпускается довольно много, в чем убедились выше).

```python
genre_profit = df_actual.pivot_table(columns='genre', index='year_of_release', 
                                     values='total_sales').reset_index(drop=False)
genre_profit.columns.name = None
genre_profit
```

```python
fig, ax = plt.subplots(figsize=(8,5))

plt.plot(genre_profit['year_of_release'], genre_profit['Shooter'], 'o-', color='gold', label='Shooter')
plt.plot(genre_profit['year_of_release'], genre_profit['Sports'], 'o-', color='green', label='Sports')
plt.plot(genre_profit['year_of_release'], genre_profit['Role-Playing'], 'o-', color='lime', label='RPG')
plt.plot(genre_profit['year_of_release'], genre_profit['Fighting'], 'o-', color='olivedrab', label='Fighting')
plt.plot(genre_profit['year_of_release'], genre_profit['Platform'], 'o-', color='blue', label='Platform')
plt.plot(genre_profit['year_of_release'], genre_profit['Action'], 'o-', color='darkorange', label='Action')
plt.plot(genre_profit['year_of_release'], genre_profit['Misc'], 'o-', color='aqua', label='Misc')
plt.plot(genre_profit['year_of_release'], genre_profit['Adventure'], 'o-', color='magenta', label='Adventure')

ax.set_ylabel('миллионы проданных копий (в среднем на игру)') 
ax.set_xlabel('год выпуска') 
ax.set_xticks(genre_profit['year_of_release'])
plt.grid(axis='x', color='0.95')
plt.legend(title='жанр:')
plt.title('Средние продажи по жанрам')
plt.show()
```

Лидер по кол-ву проданных копий на игру с заметным отрывом от остальных - жанр **Shooter**. Особенно впечатляет показатель для шутеров 2015 года выпуска - почти 2 млн копий на игру в среднем.
<br>На 2 и 3 месте в 2016-м - Sports и RPG. На последнем - Adventure (и Puzzle).

Рассмотрим теперь продажи не в среднем на игру, а суммарные:

```python
df_actual.groupby('genre')['total_sales'].sum().sort_values(ascending=False)
```

За счёт количества игр здесь на 1-м месте Action. Аутсайдер - Puzzle.

## Гипотезы
Проверим гипотезы:
* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.


Будем использовать **t-test**. Имеем дело с большими выборками, которые близки к генеральным совокупностям за 2013-2016.

Зададим уровень значимости для проверки гипотез **α=0.01**.

```python
alpha = 0.01
```

### XOne VS PC

**H<sub>0</sub>**: средние пользовательские рейтинги платформ Xbox One и PC **равны**.
<br>**H<sub>1</sub>**: средние пользовательские рейтинги платформ Xbox One и PC **различаются**.

```python
df_actual.query('platform in ["XOne","PC"]').groupby('platform')['user_score'].count(
).plot(kind='pie', ylabel='кол-во игр', figsize=(3,3));
```

Размеры совокупностей отличаются несущественно. Предположим, что стандартные отклонения одинаковы.

```python
mu_x1    = df_actual.query('platform == "XOne"')['user_score'].mean()
var_x1   = np.var(df_actual.query('platform == "XOne"')['user_score'])
sigma_x1 = np.std(df_actual.query('platform == "XOne"')['user_score'])
mu_pc    =  df_actual.query('platform == "PC"')['user_score'].mean()
var_pc   = np.var( df_actual.query('platform == "PC"')['user_score'])
sigma_pc = np.std( df_actual.query('platform == "PC"')['user_score'])
print(f"XOne: \u03BC={mu_x1:2.1f},  \u03C3\u00b2={var_x1:3.0f},  \u03C3={sigma_x1:2.1f}")
print(f"PC:   \u03BC={mu_pc:2.1f},  \u03C3\u00b2={var_pc:3.0f},  \u03C3={sigma_pc:2.1f}")
```

```python
results = st.ttest_ind(
    df_actual.query('platform == "XOne"')['user_score'], 
    df_actual.query('platform == "PC"')['user_score'],
    equal_var=True, nan_policy='omit'
)

print('p-value =', results.pvalue)

if results.pvalue < alpha:
    print('p-value < \u03B1\nОтвергаем нулевую гипотезу.\n'
          +'Делаем вывод, что средние пользовательские рейтинги платформ Xbox One и PC различаются.')
    
else:
    print('p-value \u2265 \u03B1\nНе отвергаем нулевую гипотезу.\n'
          +'Делаем вывод, что средние пользовательские рейтинги платформ Xbox One и PC равны.')
```

Полученное p-value велико, поэтому нулевую гипотезу отвергнуть не получилось бы, даже если бы α взяли больше.

```python
print('Средний рейтинг для Xbox One и PC - примерно', 
      round(df_actual.query('platform in ["XOne","PC"]')['user_score'].mean(), 1)
     )
```

### Action VS Sports

**H<sub>0</sub>**: средние пользовательские рейтинги жанров Action и Sports **равны**.
<br>**H<sub>1</sub>**: средние пользовательские рейтинги жанров Action и Sports **различаются**.

<a href="#genres">Ранее</a> видели, что кол-во игр в этих жанрах сильно отличается. Поэтому примем параметр `equal_var = False`.

Визуализируем для наглядности распределения по основным жанрам:

```python
fig, ax = plt.subplots(figsize=(14,6))
plt.title('Оценки пользователей')
sns.histplot(data=df_actual.query('genre in ["Action","Sports","Role-Playing","Shooter"]'), 
            x='user_score', stat='probability', hue='genre', common_norm=False, 
            element='poly', fill=True, kde=True);
```

```python
mu_action    = df_actual.query('genre == "Action"')['user_score'].mean()
var_action   = np.var(df_actual.query('genre == "Action"')['user_score'])
sigma_action = np.std(df_actual.query('genre == "Action"')['user_score'])
mu_sports    = df_actual.query('genre == "Sports"')['user_score'].mean()
var_sports   = np.var(df_actual.query('genre == "Sports"')['user_score'])
sigma_sports = np.std(df_actual.query('genre == "Sports"')['user_score'])
print(f"Action:  \u03BC={mu_action:2.1f},  \u03C3\u00b2={var_action:3.0f},  \u03C3={sigma_action:2.1f}")
print(f"Sports:  \u03BC={mu_sports:2.1f},  \u03C3\u00b2={var_sports:3.0f},  \u03C3={sigma_sports:2.1f}")
```

```python
results = st.ttest_ind(
    df_actual.query('genre == "Action"')['user_score'], 
    df_actual.query('genre == "Sports"')['user_score'],
    equal_var=False, nan_policy='omit'
)

print('p-value =', results.pvalue)

if results.pvalue < alpha:
    print('p-value < \u03B1\nОтвергаем нулевую гипотезу.\n'
          +'Делаем вывод, что средние пользовательские рейтинги жанров Action и Sports различаются.')
    
else:
    print('p-value \u2265 \u03B1\nНе отвергаем нулевую гипотезу.\n'
          +'Делаем вывод, что средние пользовательские рейтинги жанров Action и Sports равны.')
```

Здесь так же вывод очевиден. p-value очень мало, а средний пользовательский рейтинг для Action горзадо выше, чем для Sports:
- Action: 68.4
- Sports: 52.4
